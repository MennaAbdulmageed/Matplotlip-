#Histogram
student marks

```python
plt.hist(student_marks['mark']); #; = no read captions
```

```python
# Refer to the documentation, there are many parameters that you can use.
plt.hist(student_marks['mark'], edgecolor = 'black', bins = 11 , color = (0.5,0.1,0.5,0.5) )

#add title and axis names
plt.title('Histogram of students\'Marks' )   #note the \
plt.xlabel('Mark')
plt.ylabel('Count')

#show graph
plt.show()
```

```python
x = [0,5,9,10,15]
y = [0,1,2,3,4]
z = [4,3,2,1,0]

flg, ax = plt.subplots()
ax.plot(x,y, linestyle = "--")
ax.plot(y,z,linestyle = "-.")
flg.savefig ('linechart.jpg')

plt.show()
```

#pie chart

```python
labels = [ 'Newyork' , 'London', 'Sydeny', 'Egypt']
sizes = [50,30,45,10]

fig,ax = plt.subplots (figsize = (5,5))
ax.pie(x = sizes, labels= labels, autopct = '%.1f%%', textprops = {'fontsize':14})
ax.set_title('Distribution of Branch',fontsize = 18)

plt.show()
```

#bar chart

```python
fruits = ['apple','blueberry', 'cherry', 'orange']
votes = [ 40,100,30,55]
bar_colors = ['r','tab:blue','tab:red','tab:orange']

plt.bar(fruits,votes,color = bar_colors)
#add title and axis names
plt.title('Most loved fruits')
plt.ylabel('Votes')
plt.show()
```
